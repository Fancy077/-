```python
import requests
from bs4 import BeautifulSoup
url = "https://www.books.com.tw/web/sys_saletopb/books"
req = requests.get(url).text
soup = BeautifulSoup(req,"html.parser")
sels = soup.select("li.item")
for i in sels:
    number = i.find('strong','no').text
    title = i.find('h4').text
    price = i.select('li.price_a')
    print(number,title,price[0].text)
```
